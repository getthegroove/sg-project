# Plots der Rohdaten und der gefensterten Daten erstellen

Zunächst werden die entsprechenden Imports gemacht, die im File benötigt werden

```python
import csv 
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import tree
from sklearn.externals.six import StringIO
import pydot 
import pickle
import io
import base64
from IPython.display import HTML
%matplotlib inline  
```

Nun plotten wir zunächst die Rohdaten, die wir mithilfe einer Testperson aufgezeichnet haben. Es handelt sich bei der Bewegung um Klimmzüge, die parallel zur Aufzeichnung der Messdaten auch gefilmt wurde. Auf dem Plot lässt sich vor allem bei der Y-Achse eine wiederkehrende Struktur erkennen.

```python
X,Y,Z,temps = [],[],[],[]
with open("Dennis_chinup2.csv") as f:
  reader = csv.reader(f, delimiter=',', quotechar='|') 
  for row in reader:
      #print(row)
    temps.append(int(row[0]))
    X.append(float(row[1]))
    Y.append(float(row[2]))
    Z.append(float(row[3]))
plt.plot(temps,X,label="X")
plt.plot(temps,Y,label="Y")
plt.plot(temps,Z,label="Z")
plt.xlabel("time[ms]")
plt.ylabel("accelration[m/s^2]")
plt.legend(bbox_to_anchor=(1.00, 0.4), loc=3, borderaxespad=0.)
plt.show()    
```

# Decision-tree erstellen

Mithilfe der Zeitfenster-Datei und Zeitfernster-Dateien für andere, einfache Bewegungen haben wir nun einen decision-tree erstellen lassen. Hierbei haben wir die Dateien nacheinander eingelesen und jedes Zeitfenster seiner zugehörigen Klasse (jede Zeitfenster-Datei steht für eine Bewegung und daher für eine Klasse) zugeordnet. Zum Erstellen des decision-trees haben wir die Bibliothek sklearn verwendet. Der entstandene Baum kann unten betrachtet werden.

```python
files = []
Y = []
treename = "demo_tree"
inputFiles = ["static_stat.csv","right_left_stat.csv","up_down_stat.csv","Dennis_chinup2_stat.csv"]
i = 0
for f in inputFiles:
    with open(f,"r") as csvfile:
        tmp_list = []
        reader = csv.reader(csvfile, delimiter=',', quotechar='|')
        for row in reader:
           files = files + [row[2:8]]
           Y.append(i)
        i += 1   

clf = tree.DecisionTreeClassifier()
clf = clf.fit(files, Y)
pickle.dump(clf, open("save_"+treename+".p","wb"))

with open(treename+".dot", "w") as f:
    f = tree.export_graphviz(clf, out_file=f)

dot_data = StringIO() 
tree.export_graphviz(clf, out_file=dot_data) 
graph = pydot.graph_from_dot_data(dot_data.getvalue()) 
graph.write_pdf(treename+".pdf")

from IPython.display import HTML
HTML("<iframe src="+treename+".pdf width=100% height=750></iframe>")
```
